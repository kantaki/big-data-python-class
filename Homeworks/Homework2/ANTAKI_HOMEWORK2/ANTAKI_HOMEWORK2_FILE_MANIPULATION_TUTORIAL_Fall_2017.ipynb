# Reading from a file

https://pythonspot.com/en/read-file/

```python
import os.path

# Define a filename.
filename = "testing.txt"

if not os.path.isfile(filename):
    print 'File does not exist.'
else:
    # Open the file as f. 
    # The function readlines() reads the file.
    with open(filename) as f:
        content = f.read().splitlines()
        
    # Show the file contents line by line.
    # We added the comma to print single newlines and not double newlines.
    # This is because the lines contain the newline character '\n'. 
    for line in content:
        print(line)
```

+ ### read()		

```python
#return one big string
filename = "testing.txt"
f = open(filename, "r")
print f.read()
```

+ ### readline()

```python
#readline	#return one line at a time
filename = "testing.txt"
f = open(filename, "r")
print f.readline()
```

+ ### readlines()

```python
#returns a list of lines
filename = "testing.txt"
f = open(filename, "r")
print f.readlines()
```

# Writing on a File

```python
# Filename to write
filename = "newfile.txt"


# Open the file with writing permission
myfile = open(filename, 'w') # 'w':creates a new file (or overwrites) with the data.

# Write a line to the file
myfile.write('Written with Python\n')

# Close the file
myfile.close()
```

## Flag Actions

### Append a File

+ ##### 'a' add content to the file
+ ##### 'r'  Open file for reading 
+ ##### 'w'  Open file for writing (will truncate file) 
+ ##### 'b'  binary more 
+ ##### 'r+' open file for reading and writing 
+ ##### 'a+' open file for reading and writing (appends to end) 
+ ##### 'w+' open file for reading and writing (truncates files)
