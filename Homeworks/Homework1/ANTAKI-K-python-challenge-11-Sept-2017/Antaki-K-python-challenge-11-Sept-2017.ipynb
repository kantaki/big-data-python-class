```python
## https://stackoverflow.com/questions/6972764/python-look-and-say-sequence-improved
## 	answered Aug 7 '11 at 12:55 by agf
## Took a method that would implement the look and say algorithm and used it
def looksay(look):
    look = str(look)
    prev, count, say = look[0], 1, ''
    for char in look[1:]:
        if char == prev:
            count += 1
            continue
        say += str(count) + prev
        prev = char
        count = 1
    return say + str(count) + prev

temp = 1 ## first element of sequenc
for x in range(30):
    temp = looksay(temp)
    
print(len(temp)) ##len(a[30]) =>5808 Result
```

```python
## Antoher method Using groupby (I wasnt aware discoverd it later but it's a one liner so interesting )

from itertools import groupby
temp = 1
for x in range(30):
    temp = "".join([str(len(list(length))) + number for number,length in groupby(str(temp))])
print(len(temp)) ##-> 5808
```

```python
## and again did not figure it out
## trying to transpose the numpy array did not yield to any result
## searched the web and the solution is to shift the rows horizontally until they line up
## the Magenta Lines are the ones we want to shift
## credit where it's due : http://www.intelligentgeek.com/home/blog/pythonchallenge16
URL = 'http://www.pythonchallenge.com/pc/return/mozart.gif'
file = cStringIO.StringIO(urllib.urlopen(URL).read())
img = Image.open(file)
imgB = Image.new('RGB',img.size,(0,0,0))

def grabImgLine(src,row,start,end,inc):
    data = []
    queue = []
    offset = start
    for c in range(start,end,inc):
        val = src.getpixel( (c,row) )
        if val<240:
            queue.append(val)
        else:
            offset = c
            break
    for c in range(offset,end,inc):
        data.append(src.getpixel( (c,row) ))
    return data,queue
    
def writeImgLine(dest,row,data,queue):
    x = len(data)+len(queue)
    for c in data:
        dest.putpixel( (x,row), c )
        x -= 1
    for c in queue:
        dest.putpixel( (x,row), c )
        x -= 1
          
width,height = img.size
for y in range(0,height):
    data,queue = grabImgLine(img,y,width-1,0,-1)
    writeImgLine(imgB,y,data,queue)

imgB.save('straight.gif') ##Romance!!
```
